```python
import numpy as np 
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

```python
def ode(x, y):
  f, f_prime = y 
  f_2prime = (1 + f_prime**2) / (2*(1-f))
  return [f_prime, f_2prime]

def bcs(ya, yb):
  return [ya[0] - 1, yb[0]]

```

```python
if __name__ == "__main__":
  a = 0
  b = 1
  xs = np.linspace(a, b, 100)
  y_init = np.zeros((2, xs.size))

  f_numerical = solve_bvp(ode, bcs, xs, y_init)

  #PLOT 
  plt.plot(f_numerical.x, f_numerical.y[0], linewidth = 5, color = "pink")
  plt.xlabel("x", fontsize = 14)
  plt.grid()
  plt.ylabel(f"f(x)", fontsize = 14)
  plt.title(rf"Numerical Solution: $1+f'^2 = 2f''(y_a - f)$", fontsize = 18)
```
